```python
def columns_to_title(df):
    for column in df.columns:
        new_column = column.split("_")
        new_column = " ".join(new_column)
        new_column = new_column.title()
        df.rename(columns={column: new_column}, inplace=True)
    return df


pd_merge_test = columns_to_title(pd_merge_test)
pd_merge = columns_to_title(pd_merge)


def name_to_title(x):
    new_name = x.split("_")
    new_name = " ".join(new_name)
    new_name = new_name.title()
    return new_name
```

```python
def plot(df, title,ylabel):
    ax2 = df.plot(
        x="Epoch",
        figsize=(15, 10),
        linewidth=5,
        kind="line",
        legend=True,
        fontsize=16,
        marker="o",
        markersize=10,
    )
    ax2.legend(loc="best", prop={"size": 20})
    ax2.grid(True)
    ax2.set_title(title, fontsize=30)
    ax2.set_xlabel("Epoch", fontsize=20)
    ax2.set_ylabel(ylabel, fontsize=20)
    # ax1.xaxis.set_minor_locator(AutoMinorLocator(4))
    ax2.set_xticks(np.arange(1, 21))
    ax2.set_yticks(np.arange(0, 1.1, 0.1))
    # ax1.yaxis.set_minor_locator(AutoMinorLocator(4))
    ax2.grid(which="major", color="#CCCCCC", linestyle="--")
    ax2.grid(which="minor", color="#CCCCCC", linestyle=":")
    # ax2.figure.savefig(name + ".png")
```

```python
def acc_train_columns(df):
    return sorted(
        [
            x
            for x in df.columns
            if "Epoch" in x
            or "Train" in x
            and "Acc" in x
            and "Eval" not in x
            and "Loss" not in x
            and "Lime" not in x
            and "Aug" not in x
            and "Pretrained" not in x
            and "Resnet" not in x
        ]
    )


def acc_eval_columns(df):
    return sorted(
        [
            x
            for x in df.columns
            if "Epoch" in x
            or "Train" not in x
            and "Acc" in x
            and "Eval" in x
            and "Loss" not in x
            and "Lime" not in x
            and "Aug" not in x
            and "Pretrained" not in x
            and "Resnet" not in x
        ]
    )


def test_acc_columns(df):
    old_columns = [
        x
        for x in df.columns
        if "Epoch" in x
        or "Acc" in x
        and "Loss" not in x
        and "Time" not in x
        and "Aug" not in x
        and "Pretrained" not in x
        and "Resnet" not in x
    ]
    new_columns = [name_to_title(x) for x in old_columns]
    for i in range(len(old_columns)):
        df.rename(columns={old_columns[i]: new_columns[i]}, inplace=True)
    new_df = df[new_columns]
    return new_df


print(pd_merge[acc_train_columns(pd_merge)].max())
print(pd_merge[acc_eval_columns(pd_merge)].max())
print(test_acc_columns(pd_merge_test).max())
```
