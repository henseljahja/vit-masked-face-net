```python
pd_merge = pd.read_csv("/home/hensel/projects/vit-masked-face-net/results/train_merged_results.csv")
pd_merge_test = pd.read_csv("/home/hensel/projects/vit-masked-face-net/results/test_merged_results.csv")
```

```python
def columns_to_title(df):
    for column in df.columns:
        new_column = column.split("_")
        new_column = " ".join(new_column)
        new_column = new_column.title()
        df.rename(columns={column: new_column}, inplace=True)
    return df


pd_merge_test = columns_to_title(pd_merge_test)
pd_merge = columns_to_title(pd_merge)


def name_to_title(x):
    new_name = x.split("_")
    new_name = " ".join(new_name)
    new_name = new_name.title()
    return new_name
```

```python
def plot(df, title,ylabel):
    ax2 = df.plot(
        x="Epoch",
        figsize=(15, 10),
        linewidth=5,
        kind="line",
        legend=True,
        fontsize=16,
        marker="o",
        markersize=10,
    )
    ax2.legend(loc="best", prop={"size": 20})
    ax2.grid(True)
    ax2.set_title(title, fontsize=30)
    ax2.set_xlabel("Epoch", fontsize=20)
    ax2.set_ylabel(ylabel, fontsize=20)
    # ax1.xaxis.set_minor_locator(AutoMinorLocator(4))
    ax2.set_xticks(np.arange(1, 21))
    if "Loss" in title:
        max_val = math.ceil(df[[x for x in df.columns if "Epoch" not in x]].max().max())+0.1
        # ax2.set_yticks(np.arange(0, max_val, 0.1))
    elif "Time" in title:
        max_val = math.ceil(df[[x for x in df.columns if "Epoch" not in x]].max().max())+1
        if max_val < 5:
            max_val = 5
        # ax2.set_yticks(np.arange(0, max_val, 5))
    else:
        ax2.set_yticks(np.arange(0, 1.1, 0.1))
    # ax1.yaxis.set_minor_locator(AutoMinorLocator(4))
    ax2.grid(which="major", color="#CCCCCC", linestyle="--")
    ax2.grid(which="minor", color="#CCCCCC", linestyle=":")
    ax2.figure.savefig("../results/fig/" + title + ".png")
```

```python
def acc_train_columns(df):
    return sorted(
        [
            x
            for x in df.columns
            if "Epoch" in x
            or "Train" in x
            and "Acc" in x
            and "Eval" not in x
            and "Loss" not in x
            and "Lime" not in x
            and "Aug" not in x
            and "Pretrained" not in x
            and "Resnet" not in x
        ]
    )


def acc_eval_columns(df):
    return sorted(
        [
            x
            for x in df.columns
            if "Epoch" in x
            or "Train" not in x
            and "Acc" in x
            and "Eval" in x
            and "Loss" not in x
            and "Lime" not in x
            and "Aug" not in x
            and "Pretrained" not in x
            and "Resnet" not in x
        ]
    )


def test_acc_columns(df):
    old_columns = [
        x
        for x in df.columns
        if "Epoch" in x
        or "Acc" in x
        and "Loss" not in x
        and "Time" not in x
        and "Aug" not in x
        and "Pretrained" not in x
        and "Resnet" not in x
    ]
    new_columns = [name_to_title(x) for x in old_columns]
    for i in range(len(old_columns)):
        df.rename(columns={old_columns[i]: new_columns[i]}, inplace=True)
    new_df = df[new_columns]
    return new_df


print(pd_merge[acc_train_columns(pd_merge)].max())
print(pd_merge[acc_eval_columns(pd_merge)].max())
print(test_acc_columns(pd_merge_test).max())
```

```python
plot(pd_merge[acc_train_columns(pd_merge)],title="Train Accuracy",ylabel="Accuracy")
plot(pd_merge[acc_eval_columns(pd_merge)],"Eval Accuracy","Accuracy")
```

```python
def loss_train_columns(df):
    return sorted(
        [
            x
            for x in df.columns
            if "Epoch" in x
            or "Train" in x
            and "Acc" not in x
            and "Eval" not in x
            and "Loss" in x
            and "Lime" not in x
            and "Aug" not in x
            and "Pretrained" not in x
            and "Resnet" not in x
        ]
    )


def loss_eval_columns(df):
    return sorted(
        [
            x
            for x in df.columns
            if "Epoch" in x
            or "Train" not in x
            and "Acc" not in x
            and "Eval" in x
            and "Loss" in x
            and "Lime" not in x
            and "Aug" not in x
            and "Pretrained" not in x
            and "Resnet" not in x
        ]
    )


def loss_test_columns(df):
    old_columns = [
        x
        for x in df.columns
        if "Epoch" in x
        or "Acc" not in x
        and "Loss"  in x
        and "Time" not in x
        and "Aug" not in x
        and "Pretrained" not in x
        and "Resnet" not in x
    ]
    return sorted(old_columns)


print(pd_merge[loss_train_columns(pd_merge)].min())
print(pd_merge[loss_eval_columns(pd_merge)].min())
print(pd_merge_test[loss_test_columns(pd_merge_test)].min())
```
