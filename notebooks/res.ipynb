```python
def columns_to_title(df):
    for column in df.columns:
        new_column = column.split("_")
        new_column = " ".join(new_column)
        new_column = new_column.title()
        df.rename(columns={column: new_column}, inplace=True)
    return df
pd_merge_test = columns_to_title(pd_merge_test)
pd_merge = columns_to_title(pd_merge)

def name_to_title(x):
    new_name = x.split("_")
    new_name = " ".join(new_name)
    new_name = new_name.title()
    return new_name
```

```python
def acc_train_columns(df):
    return sorted(
        [
            x
            for x in df.columns
            if "Epoch" in x
            or "Train" in x
            and "Acc" in x
            and "Eval" not in x
            and "Loss" not in x
            and "Lime" not in x
            and "Aug" not in x
            and "Pretrained" not in x
            and "Resnet" not in x
        ]
    )


def acc_eval_columns(df):
    return sorted(
        [
            x
            for x in df.columns
            if "Epoch" in x
            or "Train" not in x
            and "Acc" in x
            and "Eval" in x
            and "Loss" not in x
            and "Lime" not in x
            and "Aug" not in x
            and "Pretrained" not in x
            and "Resnet" not in x
        ]
    )


def test_acc_columns(df):
    old_columns = [
            x
            for x in df.columns
            if "Epoch" in x
            or "Acc" in x
            and "Loss" not in x
            and "Time" not in x
            and "Aug" not in x
            and "Pretrained" not in x
            and "Resnet" not in x
        ]
    new_columns = [name_to_title(x) for x in old_columns]
    for i in range(len(old_columns)):
        df.rename(columns={old_columns[i]: new_columns[i]}, inplace=True)
    new_df = df[new_columns]
    return new_df


# train_acc = pd_merge[acc_train_columns(pd_merge)].max()
# val_acc = pd_merge[acc_eval_columns(pd_merge)].max()
test_acc = test_acc_columns(pd_merge_test).max()
test_acc
```
